## Examples of Violin Plots

```
The table modeanalytics.chick_weights contains records of 71 six-week-old baby chickens (aka chicks) 
and includes observations on their particular feed type, sex, and weight.
```

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('chick_weights_table.csv')
df.head()
````

we know the dataset doesn't need preprocessing, so we start with the violin plots

````python
sns.set(style="whitegrid")

fig, axe = plt.subplots(figsize=(8, 8))

sns.violinplot(x="feed",y="weight",data=df, inner="box", palette="Set3", cut=2, linewidth=3)

sns.despine(left=True)

fig.suptitle('Chick weights vs feed type', fontsize=18, fontweight='bold')
axe.set_xlabel("Feed type",size = 16,alpha=0.7)
axe.set_ylabel("Weight (g)",size = 16,alpha=0.7)
````

```
This violin plot shows the relationship of feed type to chick weight. The box plot elements show the median weight for horsebean-fed chicks is lower than for other feed types. The shape of the distribution (extremely skinny on each end and wide in the middle) indicates the weights of sunflower-fed chicks are highly concentrated around the median.
```

## Horizontal violin plot

````python
sns.set(style="whitegrid")

fig, axe = plt.subplots(figsize=(8, 8))
sns.violinplot(x="weight",y="feed",data=df, palette="Set3", inner="points", bw =.2, cut=2, linewidth=3)

sns.despine(left=True)

fig.suptitle('Chick weights (g) vs feed type', fontsize=18, fontweight='bold')
axe.set_xlabel("Weight (g)",size = 16,alpha=0.7)
axe.set_ylabel("Feed type",size = 16,alpha=0.7)
````

```
Horizontal violin plots are ideal for dealing with many categories. 
Swapping axes gives the category labels more room to breathe.

Points come in handy when your dataset includes observations for an entire population (rather than a select sample). 
When you have the whole population at your disposal, you don't need to draw inferences for an unobserved population; you can assess what's in front of you.

Reducing the kernel bandwidth generates lumpier plots, which can aid in identifying minor clusters, such as the tail of casein-fed chicks.
```

````python
fig, axe = plt.subplots(2,1,figsize=(24, 10))
plt.subplot(1,2,1)
plt.title('Chick weights (g) vs feed type',fontsize=35)
sns.set(style="whitegrid")
sns.violinplot(x="feed",y="weight",data=df, inner="box", palette="Set3", cut=2, linewidth=3)
sns.despine(left=True)
#------------------------------------------------------------------
plt.subplot(1,2,2)
plt.title('Chick weights (g) vs feed type',fontsize=35)
sns.set(style="whitegrid")
sns.violinplot(x="weight",y="feed",data=df, palette="Set3", inner="points", bw =.2, cut=2, linewidth=3)
sns.despine(left=True)
````

## Grouped violin plot

```
Violin plots can also illustrate a second-order categorical variable. You can create groups within each category. For instance, you can make a plot that distinguishes between male and female chicks within each feed type group.
```

````python

sns.set(style="whitegrid", palette="pastel", color_codes=True)

fig, axe = plt.subplots(figsize=(8, 8))

sns.violinplot(x="feed", y="weight", hue="sex", data=df,
               palette={"male": "b", "female": "y"})
sns.despine(left=True)

fig.suptitle('Chick weights by feed type', fontsize=18, fontweight='bold')
axe.set_xlabel("Feed",size = 16,alpha=0.7)
axe.set_ylabel("Weight (g)",size = 16,alpha=0.7)
plt.legend(loc='upper left')
````

```
The grouped violin plot shows female chicks tend to weigh less than males in each feed type category. Further, you can draw conclusions about how the sex delta varies across categories: the median weight difference is more pronounced for linseed-fed chicks than soybean-fed chicks.
```

## Grouped violin plot with split violins

````python
sns.set(style="whitegrid", palette="pastel", color_codes=True)

f, ax = plt.subplots(figsize=(8, 8))
sns.violinplot(x="feed", y="weight", hue="sex", data=df, split=True, linewidth=2.5,
               inner="quart", palette={"male": "b", "female": "y"})
sns.despine(left=True)

f.suptitle('Chick weights by feed type', fontsize=18, fontweight='bold')
ax.set_xlabel("Feed",size = 16,alpha=0.7)
ax.set_ylabel("Weight (g)",size = 16,alpha=0.7)
plt.legend(loc='upper left')
````

```
The split violins should help you compare the distributions of each group. 
For instance, you might notice that female sunflower-fed chicks have a long-tail distribution below the first quartile, 
whereas males have a long-tail above the third quartile.

In business, the term long tail is applied to rank-size distributions or rank-frequency distributions (primarily of popularity).
This is used to describe the retailing strategy of selling many unique items with relatively small quantities sold of each (the "long tail")—usually in addition to selling fewer popular items in large quantities (the "head"). 

The specific cutoff of what part of a distribution is the "long tail" is often arbitrary, but in some cases may be specified objectively.
```
